```python
df_clientes=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[0]).option("user",user).option("password",passwd).load()
df_cobertura=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[1]).option("user",user).option("password",passwd).load()
df_medio=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[2]).option("user",user).option("password",passwd).load()
df_tipo_medio=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[3]).option("user",user).option("password",passwd).load()
df_producto=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[4]).option("user",user).option("password",passwd).load()
df_tipo_producto=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[5]).option("user",user).option("password",passwd).load()
df_ejecutivo=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[6]).option("user",user).option("password",passwd).load()
df_ejecutivo_digital=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[7]).option("user",user).option("password",passwd).load()
df_grupo_laboral=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[8]).option("user",user).option("password",passwd).load()
df_contrato=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[9]).option("user",user).option("password",passwd).load()
df_orden=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[10]).option("user",user).option("password",passwd).load()
df_orden_detalle=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[11]).option("user",user).option("password",passwd).load()
df_pedido=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[12]).option("user",user).option("password",passwd).load()
df_condicion_pago=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[13]).option("user",user).option("password",passwd).load()
df_forma_factura=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[14]).option("user",user).option("password",passwd).load()
df_pedido_detalle=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[15]).option("user",user).option("password",passwd).load()
df_tipo_venta=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[16]).option("user",user).option("password",passwd).load()
df_tipo_componente=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[17]).option("user",user).option("password",passwd).load()
df_categoria_digital=spark.read.format("jdbc").option("driver",driver).option("url",url).option("dbtable",esquemasql+"."+table[18]).option("user",user).option("password",passwd).load()

```

```python
df_clientes=df_clientes.withColumn("INGESTION_DATE",current_timestamp())
df_cobertura=df_cobertura.withColumn("INGESTION_DATE",current_timestamp())
df_medio=df_medio.withColumn("INGESTION_DATE",current_timestamp())
df_tipo_medio=df_tipo_medio.withColumn("INGESTION_DATE",current_timestamp())
df_producto=df_producto.withColumn("INGESTION_DATE",current_timestamp())
df_tipo_producto=df_tipo_producto.withColumn("INGESTION_DATE",current_timestamp())
df_ejecutivo=df_ejecutivo.withColumn("INGESTION_DATE",current_timestamp())
df_ejecutivo_digital=df_ejecutivo_digital.withColumn("INGESTION_DATE",current_timestamp())
df_grupo_laboral=df_grupo_laboral.withColumn("INGESTION_DATE",current_timestamp())
df_contrato=df_contrato.withColumn("INGESTION_DATE",current_timestamp())
df_orden=df_orden.withColumn("INGESTION_DATE",current_timestamp())
df_orden_detalle=df_orden_detalle.withColumn("INGESTION_DATE",current_timestamp())
df_pedido=df_pedido.withColumn("INGESTION_DATE",current_timestamp())
df_condicion_pago=df_condicion_pago.withColumn("INGESTION_DATE",current_timestamp())
df_forma_factura=df_forma_factura.withColumn("INGESTION_DATE",current_timestamp())
df_pedido_detalle=df_pedido_detalle.withColumn("INGESTION_DATE",current_timestamp())
df_tipo_venta=df_tipo_venta.withColumn("INGESTION_DATE",current_timestamp())
df_tipo_componente=df_tipo_componente.withColumn("INGESTION_DATE",current_timestamp())
df_categoria_digital=df_categoria_digital.withColumn("INGESTION_DATE",current_timestamp())      
```

```python
#  Alinear orden de columnas en DATAFRAME
df_clientes=df_clientes.select(col("C_CLIENTE"),col("C_DOCUMENTO_IDENTIDAD"),col("C_NUMERO_DOCUMENTO"),col("C_RAZON_SOCIAL"),col("C_CODIGO_UBIGEO"),col("C_DIRECCION"),col("C_PAGINA_WEB"),col("C_CORREO_EMPRESA"),col("C_TIPO_CLIENTE"),col("C_ESTADO"),col("ingestion_date").alias("INGESTION_DATA"))

df_cobertura=df_cobertura.select(col("C_MEDIO"),col("C_COBERTURA"),col("C_DESCRIPCION"),col("C_CODIGO_UBIGEO"),col("INGESTION_DATE"))

df_medio=df_medio.select(col("C_MEDIO"),col("C_DESCRIPCION"),col("C_RAZON_SOCIAL"),col("C_TIPO_MEDIO"),col("INGESTION_DATE"))

df_tipo_medio=df_tipo_medio.select(col("C_TIPO_MEDIO"),col("C_DESCRIPCION"),col("INGESTION_DATE"))

df_producto=df_producto.select(col("C_MEDIO"),col("C_PRODUCTO_MEDIO"),col("C_TIPO_AVISO"),col("C_NOMBRE_PRODUCTO"),col("C_DESCRIPCION_PRODUCTO"),col("C_TIPO_PRODUCTO"),col("INGESTION_DATE"))

df_tipo_producto=df_tipo_producto.select(col("C_TIPO_PRODUCTO"),col("C_DESCRIPCION"),col("INGESTION_DATE"))

df_ejecutivo=df_ejecutivo.select(col("C_EJECUTIVO"),col("C_APELLIDOS"),col("C_NOMBRES"),col("C_USUARIO_EJECUTIVO_PADRE"),col("D_FECHA_INGRESO"),col("C_ACTIVO"),col("C_GRUPO_LABORAL"),col("INGESTION_DATE"))

df_ejecutivo_digital=df_ejecutivo_digital.select(col("N_EJECUTIVO_DIGITAL"),col("C_EJECUTIVO"),col("C_EJECUTIVO_DIGITAL"),col("C_ACTIVO"),("INGESTION_DATE"))

df_grupo_laboral=df_grupo_laboral.select(col("C_GRUPO_LABORAL"),col("C_NOMBRE_GRUPO"),col("C_GRUPO_VENTA"),col("INGESTION_DATE"))

df_contrato=df_contrato.select(col("C_CONTRATO"),col("C_CLIENTE_FK_CLI"),col("C_RAZON_SOCIAL"),col("C_DIRECCION"),col("C_DOCUMENTO_IDENTIDAD"),col("C_NUMERO_DOCUMENTO"),col("N_BRUTO_S"),col("N_BRUTO_D"),col("N_DESCUENTO_AGENCIA_D"),col("N_DESCUENTO_AGENCIA_S"),col("N_NETO_S"),col("N_NETO_D"),col("N_IGV_S"),col("N_IGV_D"),col("INGESTION_DATE"))

df_condicion_pago=df_condicion_pago.select(col("C_CONDICION_PAGO"),col("C_DESCRIPCION"),col("I_PARTES"),col("C_CONTADO"),col("C_CREDITO"),col("C_LETRA"),col("C_CANJE"),col("C_COMISIONABLE"),col("C_ABREVIADO"),col("INGESTION_DATE"))

df_forma_factura=df_forma_factura.select(col("C_FORMA_FACTURA"),col("C_FORMA_FACTURA_EFACT"),col("C_DESCRIPCION"),col("C_COD"),col("INGESTION_DATE"))

########## TRANSACCIONALES
df_orden=df_orden.select(col("C_ORDEN"),col("D_INICIO"),col("D_FIN"),col("C_MEDIO"),col("C_PRODUCTO_MEDIO"),col("C_COBERTURA"),col("C_EJECUTIVO"),col("N_MONTO_S"),col("N_MONTO_D"),col("N_MONTO_NETO_S"),col("N_MONTO_NETO_D"),col("N_TIPO_CAMBIO"),col("N_PORCENTAJE_APAP"),col("N_PEDIDO"),col("C_PAUTABLE"),col("INGESTION_DATE"))

df_orden_detalle=df_orden_detalle.select(col("N_PEDIDO"),col("N_PEDIDO_DETALLE"),col("C_ORDEN"),col("N_CANTIDAD_TOTAL"),col("N_MONTO_S"),col("N_MONTO_D"),col("N_MONTO_NETO_S"),col("N_MONTO_NETO_D"),col("N_MONTO_REAL_S"),col("N_MONTO_REAL_D"),col("INGESTION_DATE"))

df_pedido=df_pedido.select(col("N_PEDIDO"),col("D_INICIO"),col("D_FIN"),col("C_CLIENTE"),col("C_CONTRATO"),col("C_MONEDA"),col("N_TIPO_CAMBIO"),col("C_AGENCIA"),col("C_EJECUTIVO"),col("C_MEDIO"),col("N_TOTAL_S"),col("N_TOTAL_D"),col("C_FORMA_FACTURA"),col("C_CONDICION_PAGO"),col("INGESTION_DATE"))

df_pedido_detalle=df_pedido_detalle.select(col("N_PEDIDO"),col("N_PEDIDO_DETALLE"),col("C_MEDIO"),col("C_PRODUCTO_MEDIO"),col("C_COBERTURA"),col("C_TIPO_COMPONENTE"),col("C_CORTE"),col("C_EJECUTIVO"),col("N_TOTAL_S"),col("N_TOTAL_D"),col("N_CATEGORIA_DIGITAL"),col("INGESTION_DATE"))

df_tipo_venta=df_tipo_venta.select(col("C_TIPO_VENTA"),col("C_DESCRIPCION"),col("C_ACTIVO"),col("INGESTION_DATE"))
```

```python
df_clientes.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.cliente")
df_cobertura.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.cobertura")
df_medio.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.medio")
df_tipo_medio.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.tipo_medio")
df_producto.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.producto")
df_tipo_producto.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.tipo_producto")
df_ejecutivo.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.ejecutivo")
df_ejecutivo_digital.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.ejecutivo_digital")
df_grupo_laboral.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.grupo_laboral")
df_contrato.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.contrato")
df_orden.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.orden")
df_orden_detalle.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.orden_detalle")
df_pedido.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.pedido")
df_pedido_detalle.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.pedido_detalle")
df_condicion_pago.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.condicion_pago")
df_forma_factura.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.forma_factura")
df_tipo_venta.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.tipo_venta")
```
