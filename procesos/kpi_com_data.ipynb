```python
from pyspark.sql.functions import *
from pyspark.sql import functions as F
```

```python
df_ordenes_digitales=spark.read.format("delta").load("abfss://golden@sajhsvdesa01.dfs.core.windows.net/tables/ordenes_digitales")
```

```python
df_Ingresos_Medio_dig_mus=df_ordenes_digitales.groupBy(col("nombre_medio")).agg(sum("monto_fila_soles").alias("Ingresos")).sort(desc("Ingresos")).display()
```

```python
df_ventas_pivot_ejecom=df_ordenes_digitales.groupBy(col("segmento_eje"),F.date_format(F.col("fecha_inicio_orden"),"yyyy-MM").alias("Periodo")).agg(sum("monto_fila_soles").alias("Ingresos")).sort("segmento_eje","Periodo")
ingresos_eje_comercial=df_ventas_pivot_ejecom.groupBy(col("segmento_eje")).pivot("Periodo").agg(F.sum("Ingresos")).fillna(0)
display(ingresos_eje_comercial)
```

```python
df_ventas_digital_com=df_ordenes_digitales.groupBy(col("nombre_medio"),F.date_format(F.col("fecha_inicio_orden"),"yyyy-MM").alias("Periodo")).agg(sum("monto_fila_soles").alias("Ingresos")).sort("nombre_medio","Periodo")
ingresos_digital_comercial=df_ventas_digital_com.groupBy(col("nombre_medio")).pivot("Periodo").agg(F.sum("Ingresos")).fillna(0)
display(ingresos_digital_comercial)
```
